```python
from bs4 import BeautifulSoup 
import pandas as pd 
import json 
from glob import glob 
import os 
import time 
```

```python
file_dir = glob('../data/dmk_html/*.html')
```

```python


finalist = []
page = {}
for f in file_dir:
    name = os.path.basename(f).replace(".html", "_page")
    print(name)
    try:
        df_table = pd.read_html(f, encoding='utf-8',attrs={'class':'table table-bordered'})
        soup = BeautifulSoup(open(f), "html.parser")
        
        [district_name, from_yr, to_yr] = [t.text for t in soup.findAll('option', {"selected":"selected"})]
        print(district_name)

        print(f"Total tables: {len(df_table)}")

        data = {}
        table = {}
        for i in range(0, len(df_table)):
            #table.append(i)
            print('Starting Table: ' + str(i))
            #print(display(df_table[i]))
            table['Table_{}'.format(i)] = df_table[i].to_dict(orient='records')
            #df1['table'] = i
            #print(tabulate(df1.head()))
            #print(df1.shape)
            #print(df1.columns)
            print('-----x-----')
            data["រដ្ឋបាល ក្រុង/ស្រុក/ខណ្ឌៈ"] = district_name
            data["ចាប់ពីឆ្នាំ"] = from_yr
            data["ដល់ឆ្នាំ"] = to_yr
            if i == 0:
                data["ចំណូលប្រចាំឆ្នាំ"] = (table['Table_{}'.format(i)])
            if i == 1:
                data["ចំណាយប្រចាំឆ្នាំ"] = (table['Table_{}'.format(i)])
        finalist.append(data)
    
    except:
        print("No tables found on page " + str(f))
```

```python
with open('dmk_data.json', 'w') as outfile:
    json.dump(finalist, outfile, ensure_ascii=False, indent=4)
```
