```python
import requests
from bs4 import BeautifulSoup
import json
import copy
```

```python
communes= []
valid_ids = []

def scrape_communes_list():
    for x in range(10200, 252020):
        #print(x)
        if int(str(x)[-4:-2]) <= 15 and int(str(x)[-2:]) <= 25:
            #print("scraping " + str(x))

            f = requests.get("http://db.ncdd.gov.kh/pid/project/home/commune.castle?areaCode="+str(x))
            if f.status_code == 200:
                print(x)
                valid_ids.append(x)
                communes.append(f)
    communes_text = [json.loads(c.text) for c in communes]
    return communes_text
```

```python
def json_dump_projects_links(project_links):
    with open('projects_links.json', 'w') as outfile:
        json.dump(project_links, outfile, ensure_ascii=False, indent=4)
    with open('valid_ids.json', 'w') as outfile:
        json.dump(valid_ids, outfile, ensure_ascii=False, indent=4)
        
```

```python
def load_and_concat_json(files_list):
    output = []
    for file in files_list:
        with open(file) as json_file:
            data = json.load(json_file)
            output = output + (data)

    return output
```

```python
def get_html_for_commune(index,key):
    
    soup = BeautifulSoup(communes_text[index][key], 'html.parser')
    return soup

```

```python
def get_location_and_links(index):
    
    [province,district,commune] = get_html_for_commune(index,'Location').select('#content')[0].find('strong').get_text().split(" \\ ")
    
    projects = [] 

    for row in get_html_for_commune(index,'Projects').findAll('tr'):
        project = {}
        if 'row_p' in str(row.get('id')):
            for a in row.find_all('a'):
                if a['target'] == '_blank':
                    project['project_name'] = a.get_text()
                    project['project_link'] = a['href']
                
                if a['target'] == '_contract':
                    project['contract_id'] = a.get_text()
                    project['contract_link'] = a['href']
            
            projects.append(project)
    
    
    
    return {
        'location': {
            'province': province,
            'district': district,
            'commune': commune
        },
        'projects': projects
    }
    
```

```python
def get_project_html(n):
    for i,location in enumerate([project_links[p] for p in range(n)]):
        print(str(i)+ " " + str(location['location']))
        copied_location = copy.deepcopy(location)
        for project in copied_location['projects']:
            if 'project_link' in project.keys():
                p_html = requests.get("http://db.ncdd.gov.kh/"+project['project_link'])
                if p_html.status_code == 200:
                    project['project_html'] = get_project_data(p_html.text)

            if 'contract_link' in project.keys():
                c_html = requests.get("http://db.ncdd.gov.kh/"+project['contract_link'])
                if c_html.status_code == 200:
                    project['contract_html'] = get_contract_data(c_html.text)

        project_data.append(copied_location)
```

```python
def write_project_data():
    with open('project_data.json', 'w') as outfile:
        json.dump(project_data, outfile, ensure_ascii=False, indent=4)
```
